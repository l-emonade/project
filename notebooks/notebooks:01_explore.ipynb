```python
import pandas as pd
import matplotlib.pyplot as plt

# Fake example data
df = pd.DataFrame({
    "player": ["A", "B", "C"],
    "shots90": [2.1, 3.4, 1.2],
    "xa90": [0.12, 0.25, 0.08]
})

# Plot
ax = df.plot.scatter(x="shots90", y="xa90", s=120)
plt.title("Demo: Shots/90 vs xA/90")
plt.show()
```
